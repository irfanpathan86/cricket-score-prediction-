```python
import matplotlib.pyplot as plt

# Filter only the first inning
inning = data[data["inning"] == 56]

# Visualization
fig, axs = plt.subplots(5,1, figsize=(20, 50))
axs = axs.flatten()

axs[0].plot(inning["over"], inning["score_in_over"], color="red", marker="o", linewidth=1.5)
axs[0].grid()
axs[0].set_title("Line plot of Over vs Score in Over")
axs[0].set_xlabel("Over")
axs[0].set_ylabel("Score in Over")

axs[1].plot(inning["over"], inning["score_in_over"], color="blue", marker="o", linewidth=1.5)
axs[1].grid()
axs[1].set_title("Line plot of Over vs Total Score")
axs[1].set_xlabel("Over")
axs[1].set_ylabel("Total Score")

axs[2].plot(inning["over"], inning["run_rate"], color="green", marker="o", linewidth=1.5)
axs[2].grid()
axs[2].set_title("Line plot of Over vs Run Rate")
axs[2].set_xlabel("Over")
axs[2].set_ylabel("Run Rate")

axs[3].plot(inning["over"], inning["wicket_in_over"], color="blue", marker="o", linewidth=1.5)
axs[3].grid()
axs[3].set_title("Line plot of Over vs Wicket in Over")
axs[3].set_xlabel("Over")
axs[3].set_ylabel("Wicket in Over")

axs[4].plot(inning["over"], inning["striker_run"], color="green", marker="o", linewidth=1.5, label="Striker Run")
axs[4].plot(inning["over"], inning["non_striker_run"], color="red", marker="o", linewidth=1.5, label="Non-Striker Run")
axs[4].grid()
axs[4].set_title("Line plot of Over vs Striker & Non-Striker Runs")
axs[4].set_xlabel("Over")
axs[4].set_ylabel("Runs")
axs[4].legend()

plt.show()
```

```python
#heatmap 
inning = data[data["inning"] == 56]

num_col = ["score_in_over", "prev_total_score", "run_rate", "wicket_in_over", "striker_run", "non_striker_run", "next_over_score"]

corr=  inning[num_col].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True, cmap="coolwarm", fmt='.2f',linewidths=0.5)
plt.title("correlation of columns", fontsize=14, fontweight="bold")
plt.show()
#1 means strong relation and 0 means no relation 
```
